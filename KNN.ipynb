# K Nearest Neighbor

### Detailed K-Nearest Neighbors (KNN)

 1. **Core Concept**

**K-Nearest Neighbors (KNN):**

- **Type:** Supervised Learning algorithm.
- **Purpose:** KNN is used for both classification and regression tasks:
  - **Classification:** In classification tasks, KNN predicts the category (class) of a new data point based on the majority class among its K nearest neighbors in the feature space.
  - **Regression:** In regression tasks, KNN predicts a continuous value based on the average (or weighted average) of the values of the K nearest neighbors.
- **Nature:** Instance-based learning algorithm. This means KNN makes predictions by comparing the new data point to previously observed instances in the training set, without explicitly training a model or learning parameters. Instead, it stores the training data and makes predictions based on the proximity of new points to the stored instances.

 2. **Methodology**

**How KNN Works:**

1. **Choose K (Number of Neighbors):** 
   - The user selects the value of K, which determines how many neighbors will influence the prediction for a new data point.
   - A common practice is to experiment with different values of K to optimize performance.
   
2. **Calculate Distance:**
   - For each test point, the algorithm calculates the distance to every other point in the training set. The distance metric most commonly used is **Euclidean distance**, but other metrics like **Manhattan distance** or **Minkowski distance** can also be used depending on the problem.
   
3. **Find the K Nearest Neighbors:**
   - Once the distances are calculated, KNN identifies the **K nearest neighbors** (the K points closest to the test point).
   
4. **Prediction:**
   - **Classification:** The predicted class label is assigned by majority vote among the K nearest neighbors. For example, if 3 out of 5 neighbors belong to class A, the prediction will be class A.
   - **Regression:** The predicted value is calculated as the average (or weighted average) of the target values of the K nearest neighbors.

 Methodology of K-Nearest Neighbors (KNN)

The **K-Nearest Neighbors (KNN)** algorithm is a simple, instance-based learning algorithm used for classification and regression tasks. The methodology involves determining the K nearest data points to a query point and using those neighbors to make predictions. Below is a detailed step-by-step explanation of the methodology:

---

 1. **Choose K (Number of Neighbors)**

- **Definition:** The first step is to decide the value of **K**, which refers to the number of nearest neighbors to consider when making a prediction.
  
- **How to choose K:** 
  - **Small K:** A small K (e.g., K = 1) may lead to noisy predictions because the model may be too sensitive to outliers.
  - **Large K:** A large K (e.g., K = 15 or 20) helps to smooth out noise, but if K is too large, the model may underfit the data and fail to capture important patterns.
  - **Cross-validation:** Typically, K is chosen through experimentation, often using cross-validation, to determine which value of K gives the best performance for the given dataset.

---

 2. **Calculate Distance between the Query Point and All Training Points**

- **Definition:** To find the nearest neighbors, you need to calculate the distance between the test (query) point and every point in the training set.
  
- **Common Distance Metrics:**
  - **Euclidean Distance:** The most common distance metric used for KNN.
    
    \[
    d(x, y) = \sqrt{ \sum_{i=1}^{n} (x_i - y_i)^2 }
    \]
    
    where \( x_i \) and \( y_i \) are the feature values of two data points \( x \) and \( y \), and \( n \) is the number of features.
    
  - **Manhattan Distance:** Also called "taxicab" distance, it's the sum of the absolute differences of their coordinates:
    
    \[
    d(x, y) = \sum_{i=1}^{n} |x_i - y_i|
    \]

  - **Minkowski Distance:** A generalization of both Euclidean and Manhattan distances:
  
    \[
    d(x, y) = \left( \sum_{i=1}^{n} |x_i - y_i|^p \right)^{1/p}
    \]
    
    where \( p \) is a parameter that can be tuned (when \( p = 2 \), it becomes Euclidean distance, and \( p = 1 \) becomes Manhattan distance).

  - **Cosine Similarity:** Used particularly in text classification or when the magnitude of vectors is less important than their direction (i.e., angle between them).

- **Distance Calculation:** Once the distance metric is chosen, calculate the distance between the test point and each point in the training set.

---

 3. **Identify the K Nearest Neighbors**

- **Definition:** After calculating the distance between the query point and all training points, the next step is to **select the K nearest neighbors**. These are the K points that are closest to the query point based on the distance metric.

- **How to find nearest neighbors:**
  - Sort all training points based on their distance to the query point.
  - Select the top K training points with the smallest distance.

---

 4. **Make a Prediction Based on the Nearest Neighbors**

- **For Classification:**
  - The prediction is based on the **majority vote** among the K nearest neighbors. 
  - **Majority Voting Process:**
    - For each of the K nearest neighbors, determine the class label.
    - Assign the class label that appears most frequently among the K neighbors to the query point.

    **Example:** If K = 5, and the 5 nearest neighbors are classified as [A, A, B, A, A], the predicted class label for the query point will be **A** (since it appears 4 times).

- **For Regression:**
  - The prediction is the **average** (or weighted average) of the target values of the K nearest neighbors.
  - **Averaging Process:**
    - Calculate the average of the values of the K nearest neighbors to predict the value for the query point.

    **Example:** If K = 3 and the target values of the nearest neighbors are [10, 12, 15], the predicted value will be:

    \[
    \text{Predicted Value} = \frac{10 + 12 + 15}{3} = 12.33
    \]

  - **Weighted Average:** Optionally, a weighted average can be used, where closer neighbors have more influence on the prediction. This is often done by assigning a weight inversely proportional to the distance.

---

 5. **Handle Ties (in Classification)**

- **Tie in Majority Voting:** If there is a tie in the number of nearest neighbors from different classes, there are various strategies to break the tie:
  - **Increase K:** One simple method is to increase K by 1, which may help to resolve ties.
  - **Weighted Voting:** Instead of a simple majority vote, use a **distance-weighted voting** mechanism where closer neighbors have a higher voting weight.
  - **Random Selection:** In cases where there's still a tie, some systems will choose the class randomly from the tied classes.
  
---

 6. **Return the Prediction**

- After identifying the K nearest neighbors and applying the appropriate rule (majority vote for classification or average for regression), the algorithm returns the predicted class label (for classification) or continuous value (for regression) for the query point.

---

 Example Workflow of KNN

1. **Choose K:** Suppose K = 3.
2. **Calculate Distances:** Calculate the distance from the query point to all training points using Euclidean distance.
3. **Find K Nearest Neighbors:** Sort all training points by distance, and select the 3 nearest neighbors.
4. **Make Prediction:**
   - **For Classification:** Suppose the classes of the 3 nearest neighbors are [A, A, B]. The predicted class will be **A** (since it has 2 votes).
   - **For Regression:** Suppose the target values of the 3 nearest neighbors are [5, 7, 6]. The predicted value will be the average: \( \frac{5 + 7 + 6}{3} = 6 \).
5. **Return Prediction:** The algorithm outputs the predicted class (for classification) or value (for regression).

---

 Additional Considerations

- **Normalization/Standardization:** Since KNN relies on distance calculations, it's crucial to normalize or standardize the feature values, especially when features have different units (e.g., height in centimeters and weight in kilograms). This ensures that features contribute equally to the distance computation.
  
- **Curse of Dimensionality:** As the number of features (dimensions) increases, the notion of "nearness" becomes less meaningful. This can degrade the performance of KNN, making dimensionality reduction techniques (like PCA) useful in high-dimensional spaces.

---

 Summary of KNN Methodology:

1. **Choose K**, the number of neighbors to consider.
2. **Calculate distances** between the query point and all training points.
3. **Identify the K nearest neighbors** based on the calculated distances.
4. **Make a prediction** using majority voting (for classification) or averaging (for regression).
5. **Return the prediction** for the query point.

The algorithm is simple, easy to understand, and effective for many classification and regression problems, particularly when the relationship between the data points is intuitive and not overly complex. However, it can become computationally expensive as the size of the dataset grows.

## Breast Cancer Dataset

```python
from sklearn.impute import SimpleImputer
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
cancer = pd.read_csv("./Cases/Wisconsin/BreastCancer.csv", index_col=0)
cancer.head()
```

```python
X = cancer.drop("Class", axis=1)
y = cancer['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
pd.crosstab(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
X = cancer.drop("Class", axis=1)
y = cancer['Class']
n_values = np.arange(1,11)
accuracy = []
for i in n_values:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    print(f"for nearest neighbour {i}, accuracy : {accuracy_score(y_test, y_pred)}")
    accuracy.append({'NN':i, 'accuracy': accuracy_score(y_test, y_pred)})
    
```

```python
df = pd.DataFrame(data=accuracy)
df = df.sort_values(by='accuracy', ascending=False)
df.head(1)
```
